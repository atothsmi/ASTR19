# Central Limit Theorem Demonstation

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n_exp = 10000 #Number of experiments
n_draw = 5 #Number of random variates per experiment

means = np.zeros(n_exp) #Means from each experiment
```

# Let's perform the experiments with the Uniform random distribution

For a uniform distribution between $a$ and $b$, the theoretical mean is $(b-a)/2$ and the theoretical variance is $(b-a)^2/12$ (stdev is square root of variance)

Therefore, according to the CLT the distribution of means (for our uniform distribution between $a=0$ and $b=1$) must tend to an average of $(b-a)/2 = .5$ and a stdev $(b-a)/\sqrt{12*n_{draw}}$.

```python
#loop over the number of experiments
for i in range(n_exp):
    #pull n_draw random variates from a uniform distribution
    z= np.random.uniform(0., 1., n_draw)

    #compute the mean
    means[i] = np.mean(z)
```

Define a function to plot a Guassian

```python
def gaussian(x, mu, sigma):
    return 1./np.sqrt(2*np.pi*sigma**2) * np.exp(-0.5*(x-mu)**2/sigma**2)
```

Load our matplotlib defaults

```python
plt.style.use("..\class14\\astr19-matplotlib-defaults.txt")
```

```python
fig = plt.figure(figsize=(7,7))

n_bins = 50
x_min = -0.1
x_max = 1.1

plt.hist(means, bins = n_bins, range=[x_min, x_max], alpha=0.5, edgecolor='white', density=True)

#add the gaussian from the CLT
mu = 0.5
sigma = np.sqrt(1.0/12.0) * 1.0/np.sqrt(n_draw)

xx = np.linspace(x_min, x_max, 1000)
yy = gaussian(xx, mu, sigma)
plt.plot(xx, yy, color='red')
plt.xlabel('x')
plt.ylabel('y(x)')

plt.show()
```

# let's repeat but with a coin flip

#### How to simulate a coin flip?

The randint() Numpy function set to a max of 2 (exclusive) will pull 0 or 1, which we can then use as a coin flip. 

For this distribution, the theoretical mean is 0.5, and the theoretical stdev 0.5 (variance = $0.5^2 = 0.25$) Therefore, according to the CLT the means will have a mean of 0.5, and a stdev of $0.5/\sqrt{n_{draw}}$.

```python
#change the number of draws
n_exp = 100000
n_draw = 1000
means = np.zeros(n_exp)

#loop over the number of experiments
for i in range(n_exp):
    #flip the coin n_draw times
    z= np.random.randint(0, 2, n_draw)

    #compute the mean
    means[i] = np.mean(z)
```

```python
fig = plt.figure(figsize=(7,7))

n_bins = 50
x_min = -0.1
x_max = 1.1

plt.hist(means, bins = n_bins, range=[x_min, x_max], alpha=0.5, edgecolor='white', density=True)

#add the gaussian from the CLT
mu = 0.5
sigma = 0.5 * 1.0/np.sqrt(n_draw)

xx = np.linspace(x_min, x_max, 1000)
yy = gaussian(xx, mu, sigma)
plt.plot(xx, yy, color='red')
plt.xlabel('x')
plt.ylabel('y(x)')

plt.show()
```

#### Let's repeat but with a 20-sided die

We generate values between 1 and 20, with uniform probability for each.

Now, the theoretical mean is 10.5. and the theoretical stdev $\sqrt{n_{sides}^2-1)/12} = 5.677$ (variance = 33.25). Therefore, according to the CLT, the means will have a mean of 10.5, and stdev of $5.677/\sqrt{n_{draw}}$.

```python
#change the number of draws
n_exp = 100000
n_draw = 100
means = np.zeros(n_exp)

#loop over the number of experiments
for i in range(n_exp):
    #flip the coin n_draw times
    z= np.random.randint(1, 21, n_draw)

    #compute the mean
    means[i] = np.mean(z)
```

```python
fig = plt.figure(figsize=(7,7))

n_bins = 100
x_min = 0
x_max = 21

plt.hist(means, bins = n_bins, range=[x_min, x_max], alpha=0.5, edgecolor='white', density=True)

#add the gaussian from the CLT
mu = 10.5
sigma = 5.677 * 1.0/np.sqrt(n_draw)

xx = np.linspace(x_min, x_max, 1000)
yy = gaussian(xx, mu, sigma)
plt.plot(xx, yy, color='red')
plt.xlabel('x')
plt.ylabel('y(x)')

plt.show()
```
