# Plot with Polar Coordinates

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#when angles are measured in radians, 1 turn is 2pi radians

n = 100

theta = np.linspace(0, 2*np.pi, 50)
y1 = np.ones(50)
y2 = theta*0.25

#plotting
f = plt.figure(figsize=(8,6))
ax = f.add_subplot(111, projection = 'polar')

ax.plot(theta, y1, label='constant')
ax.plot(theta, y2, label='linear')

ax.grid(True)

plt.show()
```
