Select your favorite distribution (other than uniform) and pull 1000 random numbers from that distribution. Histogram the random numbers into 100 bins, and plot the histogram. Label your axes and save the figure as a PDF.

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
#Number of elements
n = 1000

#Get a array of random numbers from a gumbel distribution
gumbel_array = np.random.gumbel(.27, .1, n)

#create 100 bins
bins = np.linspace(0, 1, 100)

plt.hist(gumbel_array, bins =bins)

plt.xlabel("Gumbel Distributed Random Numbers")
plt.ylabel("Number of Numbers per Bin")

plt.plot()

plt.savefig("GumbelDist.pdf")
```
