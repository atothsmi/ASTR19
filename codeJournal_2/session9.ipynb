 Session 9 Prompt: Create a Jupyter notebook, import matplotlib. Use numpy to pull 1000 random numbers distributed uniformly between [0,1]. Histogram the random numbers into 100 bins, and plot the histogram. Label your axes and save the figure as a PDF

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
#Number of elements
n = 1000

#Get a array of random numbers distributed uniformly
uniform_array = np.random.uniform(0.0, 1.0, n)

#create 100 bins
bins = np.linspace(0, 1, 100)

plt.hist(uniform_array, bins =bins)

plt.xlabel("Uniformly Distributed Random Numbers")
plt.ylabel("Number of Numbers per Bin")

plt.plot()

plt.savefig("rNumDist.pdf")
```
