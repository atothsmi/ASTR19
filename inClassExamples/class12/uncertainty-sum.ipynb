# Uncertainty of the Sum of Two Variables

```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
%matplotlib inline
```

```python
#fix the random seed
np.random.seed(42)

#generate data
n = 50000 #size of arrays
x = np.random.uniform(-0.5, 0.5, n) #samples = 0 +- 0.5
y = np.random.uniform(-0.5, 0.5, n) #samples = 0 +- 0.5
z = x + y

#plot histogram of data in 3 different panels
width = 0.05
my_bins = np.arange(-1.5, 1.5, width)

fig = plt.figure(figsize=(5,5))
axx = fig.add_subplot(3, 1, 1) #subplot for x histogram
axy = fig.add_subplot(3, 1, 2) #subplot for y histogram
axz = fig.add_subplot(3, 1, 3) #subplot for z histogram

axx.hist(x, bins=my_bins, alpha=0.5, edgecolor='black')
axx.set_xticks([])
axx.set_ylabel('x frequency')

axy.hist(y, bins=my_bins, alpha=0.5, edgecolor='black')
axy.set_xticks([])
axy.set_ylabel('y frequency')

axz.hist(z, bins=my_bins, alpha=0.5, edgecolor='black')
# axz.set_xticks([])
axz.set_ylabel('z frequency')

plt.subplots_adjust(hspace=0.0)

plt.show()

#statistical analysis
rows = ['x', 'y', 'z']
cols = ['mean', 'std', 'var']

data = [[np.mean(x), np.std(x), np.var(x)],
        [np.mean(y), np.std(y), np.var(y)],
        [np.mean(z), np.std(z), np.var(z)]]

table = tabulate(data, headers =cols, showindex=rows, tablefmt='fancy_grid')
print(table)
```
