# Plot with Polar Coordinates

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Shape of the dipolar field emitted by an antenna
def dipole_field(theta):
    return np.sin(theta)**2
```

```python
#when angles are measured in radians, 1 turn is 2pi radians

n = 100

#Define theta values
theta = np.linspace(0, 2*np.pi, 50)
y1 = np.ones(50)
y2 = theta*0.25

#plotting
f = plt.figure(figsize=(8,6))
ax = f.add_subplot(111, projection = 'polar')
ax.plot(theta, y1, label='constant')
ax.plot(theta, y2, label='linear')
ax.plot(theta, dipole_field(theta), label='dipolar field')

#customizing the plot
ax.set_theta_zero_location('N') #Set zero degrees to the top (North)
ax.set_theta_direction(-1) #Reverse the theta direction (clockwise)
ax.set_rlabel_position(90) #Set radial labels to apppear on the right

#add grid
ax.grid(True)

#add legent
ax.legend(loc='upper right')

#add a vertical line at the center of the plot
# specify the length of the vertical line at the center of the plot
line_length = 0.3 #adjust this value to change the length of the line
ax.plot([0, np.pi], [line_length, line_length], color='r', linestyle='-', linewidth=5)

plt.show()
```
