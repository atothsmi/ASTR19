# Make a simple plot with errors

randomly draw some numbers x and y using numpy, and then use matplotlib to plot them with error bars


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#generate some fake data
x = np.arange(10) + 1*np.random.randn(10)
y = np.arange(10) + 1*np.random.randn(10)

xerror = 1*np.random.random(10)
yerror = 1*np.random.random(10)

plt.errorbar(x, y, xerr=xerror, yerr=yerror, fmt='bo')
plt.show()
```

```python

#normal (gaussian) distribution of mean 0 and variance 1 (typical deviation from mean)
print(np.random.randn(10))

#uniform distribution between [0, 1)
print(np.random.random(10))
```
