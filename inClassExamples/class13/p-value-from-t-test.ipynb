# Computing p-value from t-test

Write a function that returns the p-value between two samples corresponding to the null hypothesis that both samples are from distributions with the same mean value

Then show two examples where we expect the p-values to be high or low.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind 
```

first write a function to return the p-value between two distributions using the t-test

```python
#function to compute p-value
def pvalue(x0, x1):
    #call the ttest_ind function from scipy.stats
    _ , p = ttest_ind(x0, x1, equal_var=False)

    #return the p-value
    return p
```

Generate two similar distributions, histogram them, and calculate p-value

```python
#number of samples
n = 1000

#first set of random variates
x0= np.random.normal(loc=0, scale=1.0, size=n)

#second set of random variates
x1= np.random.normal(loc=0, scale=1.0, size=n)

#create two histograms on the same panel
f, ax = plt.subplots (1, 1, figsize=(7,7))

ax.xaxis.set_tick_params(which = 'both', direction='in')
ax.yaxis.set_tick_params(which = 'both', direction='in')

ax.hist(x0, bins = 10, alpha=0.5, label= 'Set A')
ax.hist(x1, bins = 10, alpha=0.5, label= 'Set B')

plt.legend(frameon=False)

#compute the p-value
p = pvalue(x0, x1)

#print the result
print(f'The p-value for the null hypothesis that x0 and x1 are pulled from the same distribution is p = {p:5.4e}')

plt.show()
```

Generate two different distributions, histogram them, and calculate p-value

```python
#number of samples
n = 1000

#first set of random variates
x0= np.random.normal(loc=0, scale=1.0, size=n)

#second set of random variates
x1= np.random.normal(loc=0.5, scale=1.0, size=n)

#create two histograms on the same panel
f, ax = plt.subplots (1, 1, figsize=(7,7))

ax.xaxis.set_tick_params(which = 'both', direction='in')
ax.yaxis.set_tick_params(which = 'both', direction='in')

ax.hist(x0, bins = 10, alpha=0.5, label= 'Set A')
ax.hist(x1, bins = 10, alpha=0.5, label= 'Set B')

plt.legend(frameon=False)

#compute the p-value
p = pvalue(x0, x1)

#print the result
print(f'The p-value for the null hypothesis that x0 and x1 are pulled from the same distribution is p = {p:5.4e}')

plt.show()
```
