# Make a Figure with Geometric shapes

```python
import matplotlib.pyplot as plt
from matplotlib import patches
%matplotlib inline
```

```python
#create a figure
f = plt.figure(figsize = (6, 6))

#add a single subplpot to the figure
ax = f.add_subplot(111)

#remove x and y ticklabels

#create a rectangle
rectangle = patches.Rectangle((0.05, 0.05), 0.4, 0.4, edgecolor = 'None', facecolor='blue', alpha = 0.5)
ax.add_patch(rectangle)

#create a circle
circle = patches.Circle((0.6, 0.6), 0.3, edgecolor='None', facecolor='red', alpha = 0.5)
ax.add_patch(circle)

#create an ellipse
ellipse = patches.Ellipse((0.8, 0.8), 0.3, 0.4, edgecolor = 'None', facecolor = 'green', alpha = 0.5)
ax.add_patch(ellipse)

plt.show()
```

```python
#check if a number is odd
def is_odd(num):
    return num%2 !=0
```

```python
#create a new figure
f = plt.figure(figsize=(6,6))
ax = f.add_subplot(111)

#iterate over columns and rows to create black rectangles
for i in range(8):
    for j in range(8):
        if is_odd(i + j + 1):
            rectangle = patches.Rectangle((i, j), 1, 1, edgecolor = 'None', facecolor='k', alpha = 1)
            ax.add_patch(rectangle)

#remove ticks
ax.set_xticks([])
ax.set_yticks([])

#show on axes from 0 to 8
ax.set_xlim(0, 8)
ax.set_ylim(0, 8)

plt.show()
```
