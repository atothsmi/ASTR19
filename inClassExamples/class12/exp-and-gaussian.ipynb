# Exponential and Gaussian

```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
%matplotlib inline

#create array from 0 to 10
x = np.arange(0, 11, 1)

#create arrays for 2^x, e^x, 10^x
y1 = 2**x
y2 = np.exp(x)
y3 = 10**x

#create a table
data = list(zip(x, y1, y2, y3))
cols = ['x', '2^x', 'e^x', '10^x']

#print table
table = tabulate(data, headers=cols, tablefmt='fancy_grid')
print(table)
```

```python
#plot the 3 functions
n = 1000
x = np.linspace(-3, 3, n)

#create arrays for 2^x, e^x, 10^x
y1 = 2**x
y2 = np.exp(x)
y3 = 10**x

plt.plot(x, y1, label='2^x')
plt.plot(x, y2, label='e^x')
plt.plot(x, y3, label='10^x')
plt.ylim(0, 10)
plt.legend()

plt.show()
```

```python
x = np.linspace(-3, 3, 1000)
mean = 1.0
sigma = .5
y = np.exp(-(x-mean)**2 / (2.0 * sigma**2)) * (1 / np.sqrt(2.0 * np.pi * sigma**2))
plt.plot(x, y)
plt.show()
```

```python
def gaussian(x, mean, sigma):
    return np.exp(-(x-mean)**2 / (2.0 * sigma**2)) * (1 / np.sqrt(2.0 * np.pi * sigma**2))

#create 3 gaussian functions
x = np.linspace(-10, 10, 1000)
y1 = gaussian(x, 0, 1)
y2 = gaussian(x, 1, 2)
y3 = gaussian(x, 2, .5)

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```

How does the gaussian function compare to the normal distribution we made earlier?

```python
#generate some random data
n = 50000
mean = 0.0
sigma = 0.25
samples = np.random.normal(mean, sigma, n)

#create histogram
width = 0.05
my_bins = np.arange(-1.5, 1.5, width)
x = np.linspace(-1.5, 1.5, 1000)

#pdf = probability density function
pdf = gaussian(x, mean, sigma)

plt.hist(samples, bins=my_bins, alpha=0.5, edgecolor='black', label='histogram', density='True')
plt.plot(x, pdf, label='density function') #can also normalize by pdf*n*width
plt.xlabel('x')
plt.ylabel('y frequency')
plt.legend()

plt.show()
```
